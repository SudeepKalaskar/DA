```python
import seaborn as sns
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

```python
weight = [62,64, 69, 75, 65, 68, 65, 71, 76, 73]
height = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
```

```python
%matplotlib inline
plt.figure(figsize=(4.5,2.5))
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
sns.scatterplot(x = height, y = weight)
plt.show()
```

```python
gender = ["Female","Female","Female","Female",
          "Male","Male","Male","Male","Male","Male","Male"]
```

```python
plt.figure(figsize=(4.5,3))
plt.title("Count of Male amd female")
sns.countplot(x=gender, hue=gender,legend=["Female","Male"],fill=False)
plt.show()
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("SampleSuperstore.xlsx")
df.head()
```

```python
plt.figure(figsize=(11, 4))
sns.countplot(data=df,x="Ship Mode", hue="Segment")
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(4.5, 3))
sns.scatterplot(data=df, x="Quantity", y="Sales",hue="Category")
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.histplot(data=df, x="Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(4.5, 3))
sns.scatterplot(data=df, x="Quantity", y="Sales",hue="Category",
                hue_order=["Technology","office Supplies","Furniture"],
                palette={"Technology":"#764431", "office Supplies":"#858223",
                "Furniture":"#D42343"})
plt.show()
```
