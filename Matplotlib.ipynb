```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(4, 15, 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y, label="Sin Wave", linestyle="-",linewidth=2, color="#123457")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Waves")
plt.legend()
plt.grid(True)
plt.show()
```

```python
category = ["A", "B", "C", "D", "E"]
value = [33, 54, 42, 51, 29]
```

```python
plt.figure(figsize=(5, 3))
plt.bar(category, value, alpha=0.7, color=["#111040","#111100","#777060","#128777","#234250"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categories")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y, c=color, s=size, alpha=0.5,cmap="viridis")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.colorbar(label="color Intensity")
plt.grid(True)
plt.show()
```

```python
dice = np.array([1, 2, 3, 4, 5, 6]) # Dice
sample = np.random.choice(dice, 4) # Rolling dice 4 time in a row and getting values
print("output of 4 samples", sample)
print("Mean of above 4 samples", sample.mean())
```

```python
samples_1000 = [np.random.choice(dice, 4).mean() for i in range(1000)]
samples_1000    
```

```python
plt.figure(figsize=(7,4))
plt.hist(samples_1000, alpha=0.7, edgecolor="black", bins=1000)
plt.show()
```

```python
data = np.random.randn(10000)

plt.figure(figsize=(5,3))
plt.hist(data, alpha=0.7, edgecolor="black", bins=100, histtype="stepfilled", color="Indigo")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

```python
sizes = np.random.randint(50, 70, 4)
labels = ["East","West","North","South"]

plt.figure(figsize=(5, 4))
plt.pie(sizes, labels=labels, colors=["gold","orange","lightblue","lightcoral"],
        autopct="%1.1f%%", explode=[0.01, 0.1, 0.01, 0.01])
plt.title("Pie Chart")
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1,size=(100,))

plt.figure(figsize=(5,3))
plt.boxplot(data, patch_artist=True, vert=False)
plt.title("Boxplot")
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize=(8,6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, color="blue", linewidth=0.2)
ax.plot(angles, radii, color="blue", alpha=0.3)
plt.title("Polar Plot")
plt.show()
```

```python
matrix = np.random.rand(5,5)
plt.figure(figsize=(5, 4))
plt.imshow(matrix, cmap="plasma")
plt.title("Heatmap")
plt.colorbar(label="Color Intensity")
plt.show()
```
